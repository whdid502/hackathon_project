```python
# 날씨 데이터
date_to_weather = {}
weather_to_num = {
    '맑음' : 0,
    '구름조금' : 1,
    '구름많음' : 2,
    '흐림' : 3,
    '비' : 4,
    '눈' : 5,
}
tmp = pd.read_csv('./data/weather_2020.csv', encoding='euc-kr', header=None)

for i in range(tmp.count()[0]) :
    date_to_weather[tmp[0][i]] = weather_to_num[str(tmp[1][i]).strip()]
    
date_to_weather
```

```python
df = pd.read_csv('서울교통공사_역별 일별 시간대별 승하차인원 정보_20200831.csv', encoding='euc-kr')
times = df.columns[5:]
new_col = {}
hour = 5
for t in times :
    new_col[t] = hour
    hour += 1
df.rename(columns=new_col, inplace=True)
df
```

```python
new_df = pd.DataFrame(columns=['연도', '월', '일', '요일', '시간', '날씨', '역명', '인원수'])
times = df.columns[5:]

for i in range(df.count()['날짜']) :
    y = int(df['날짜'][i].split('-')[0])
    m = int(df['날짜'][i].split('-')[1])
    d = int(df['날짜'][i].split('-')[2])
    week = datetime.datetime(y,m,d).weekday()
    
    if week < 5 and (df['날짜'][i] in indep) :
        week = 7
        
    station = df['역명'][i]
    weather = date_to_weather[df['날짜'][i]]
    
    for t in times :
        new_df = new_df.append({'연도' : y,
                                '월' : m,
                                '일' : d,
                                '요일' : week,
                                '시간' : t,
                                '날씨' : weather,
                                '역명' : station,
                                '인원수' : df[t][i]
                               }, ignore_index=True)
        
    if i % 100 == 0 and i > 0 :
        print(f"[INFO] done {i} / {df.count()['날짜']}")
        
new_df
```

```python
new_df.groupby(['연도', '월', '일', '요일', '시간', '날씨','역명']).sum()
new_df.to_csv('new_2020.csv', encoding='euc-kr')
new_df
```
