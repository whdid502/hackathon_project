```python
import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

from sklearn.externals import joblib
```

```python
# 경고메시지 무시
import warnings
warnings.filterwarnings("ignore")
```

```python
def training(station, df, hour):
    tmp = df[df['역명'] == station].reset_index(drop=True)
#     tmp = df[hour].reset_index(drop=True)
    tmp = tmp[['연도', '월', '일', '역명', hour, '요일', '날씨']]
    tmp = pd.get_dummies(tmp, columns=['요일', '날씨'])
    
    Y = tmp[hour]
    del tmp[hour]

    del tmp['역명']
    X = tmp
    
    display(X)
    
    model = LinearRegression()

    X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=0)
    model.fit(X_train, y_train)
    
    print(f'{station}_{hour} score : {model.score(X_test, y_test)}')
    
    joblib.dump(model, f'./models/{station}_{hour}.pkl') 
```

```python
df = pd.read_csv('서울교통공사_역별 일별 시간대별 승하차인원 정보_20200831_수정.csv', encoding='euc-kr')
# del df['Unnamed: 0']
times = df.columns[4:24]
new_col = {}
hour = 5
for t in times :
    new_col[t] = hour
    hour += 1
df.rename(columns=new_col, inplace=True)
df
```

```python
station_list = df['역명'].unique()

for station in station_list :
    for hour in range(5,25):
        training(station, df, hour)
    break
```
